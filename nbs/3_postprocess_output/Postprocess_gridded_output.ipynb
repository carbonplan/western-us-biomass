```python
import matplotlib.pyplot as plt
import xarray as xr

from conus_biomass import dir_info
from conus_biomass.process_outputs import postprocess_output
```

# Do postprocessing

```python
postprocess_output.process_model_output()
```

# Look at raw output data

```python
ds = xr.open_mfdataset(
    dir_info.dir_model_output + "predicted_biomass" + "_unfiltered_" + str(2014) + "_*_.nc",
    combine="by_coords",
)
ds2 = xr.open_mfdataset(
    dir_info.dir_model_output + "predicted_biomass" + "_unfiltered_" + str(2020) + "_*_.nc",
    combine="by_coords",
)
```

```python
ds["predicted_biomass"].coarsen(x=10, y=10, boundary="trim").mean().plot(vmax=200)
```

```python
(ds2["predicted_biomass"] - ds["predicted_biomass"]).plot(vmax=15, cmap=plt.cm.BrBG)
```
