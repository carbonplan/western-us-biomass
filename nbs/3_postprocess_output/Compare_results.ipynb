```python
tseries = []
for year in np.arange(2005, 2022):
    ds = xr.open_mfdataset(
        "/dfs10/jranders_lab1/users/czarakas/uncertain_land_sink_data/processed_data/model_results/1000m_2026feb02_processed/"
        + "predicted_biomass"
        + "_filtered_"
        + str(year)
        + "_0000"
        + ".nc"
    )
    tseries.append(np.nanmean(ds["predicted_biomass"]))
```
